### Getting Mood Playlist Data

```python
master_song_list = []

for i in range(len(only_playlist_ids)):
    
    # returns playlist json file
    responses = playlist_calls(only_playlist_ids[i], headers)
    
    #get only the metrics we want
    name = responses['name']

    for i in range(len(responses['tracks']['items'])):
        song_name = responses['tracks']['items'][i]['track']['name']
        song_id = responses['tracks']['items'][i]['track']['id']

        artists = []

        for j in range(len(responses['tracks']['items'][i]['track']['artists'])):
            artists.append(responses['tracks']['items'][i]['track']['artists'][j]['name'])

    # add the playlist, song_id, song_name and artist to a list of lists. each list within the main will be a song
        master_song_list.append([name, song_id, song_name, artists])

#creating dataframe with data. each row is a song
playlist_data = pd.DataFrame(master_song_list,
                             columns=['playlist_name', 'song_id', 'song_name', 'song_artists'])


#saving as csv
playlist_data.to_csv('playlist_data.csv', index=False)    
    
```

```python
## create a list of the song ids from all my playlists in order to later get the list of playlists associated with each song

song_list = []

for song_id in playlist_data['song_id']:
    if song_id not in song_list:
        song_list.append(song_id)
```

```python
## using the song_ids from the list above, we grab the list of playlists per song

song_playlist_list = []

for song_id in song_list: 
    
    subset_data = playlist_data[playlist_data['song_id']==song_id]
    
    playlist_names_list = []
    
    for i in range(len(subset_data['playlist_name'].values)):
        
        playlist_names_list.append(subset_data['playlist_name'].values[i])
    
    song_playlist_list.append([song_id, playlist_names_list])
```

```python
import time

missing_song_features = []

## for each song in missing songs list which is the difference between the tracks df and playlist df 
for song in missing_songs:
    
    ## get audio features of each mising song
    params = (('ids', song),)

    response = requests.get('https://api.spotify.com/v1/audio-features', headers=headers, params=params)

    responses = response.json()['audio_features']

    time.sleep(0.2) 
    
    ## put response in a list for all songs
    missing_song_features.append(responses)
```

```python
missing_song_info = []

for song in missing_songs:
    
    ## get track info of each mising song
    params = (('ids', song),)

    response = requests.get('https://api.spotify.com/v1/tracks', headers=headers, params=params)

    responses = response.json()['tracks'][0]

    time.sleep(0.2) 
    
    ## put response in a list for all songs
    missing_song_info.append(responses)
```

```python
## puts track info into a list so we can put it in a dataframe

missing_song_info_list = []

for i in range(len(missing_song_info)):

    missing_song_info_dict = {'id': missing_song_info[i]['id'],
                                  'name': missing_song_info[i]['name'], 
                                   'popularity': missing_song_info[i]['popularity'],
                                   'explicit': missing_song_info[i]['explicit'],
                                   'release_datetime': missing_song_info[i]['album']['release_date']
                                 }
    artist_list = []
    
    for j in range(len(missing_song_info[i]['artists'])):

        artist_list.append(missing_song_info[i]['artists'][j]['name'])

        missing_song_info_dict['artists'] = artist_list

        
    missing_song_info_list.append(missing_song_info_dict)
```
